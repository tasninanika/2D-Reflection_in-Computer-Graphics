```python
import matplotlib.pyplot as plt

# Step 1: Original square points
square = [(1, 1), (1, 4), (4, 4), (4, 1)]

# Step 2: Reflect over X-axis (change y → -y)
reflected_x = [(x, -y) for x, y in square]

# Step 3: Reflect over Y-axis (change x → -x)
reflected_y = [(-x, y) for x, y in square]

# Step 4: Close all shapes
square.append(square[0])
reflected_x.append(reflected_x[0])
reflected_y.append(reflected_y[0])

# Step 5: Separate x and y for plotting
ox, oy = zip(*square)
rx, ry = zip(*reflected_x)
yx, yy = zip(*reflected_y)

# Step 6: Plot everything
plt.plot(ox, oy, 'go-', label='Original Square')
plt.plot(rx, ry, 'o--', label='Reflected over X-axis')
plt.plot(yx, yy, 'ro--', label='Reflected over Y-axis')

# Extras
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D Reflection of a Square")
plt.legend()
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to reflect a point about the X-axis (y becomes negative)
def reflect_x(point):
    x, y = point
    return x, -y

# Function to reflect a point about the Y-axis (x becomes negative)
def reflect_y(point):
    x, y = point
    return -x, y

# Function to reflect a point about the line y = x (swap x and y)
def reflect_xy(point):
    x, y = point
    return y, x

# Function to reflect a polygon (square in this case)
def reflect_polygon(polygon, reflection_type):
    if reflection_type == "x":
        return [reflect_x(p) for p in polygon]
    elif reflection_type == "y":
        return [reflect_y(p) for p in polygon]
    elif reflection_type == "xy":
        return [reflect_xy(p) for p in polygon]
    else:
        return polygon  # No reflection if invalid type

# Define original square and reflection type
square = [(1, 1), (1, 4), (4, 4), (4, 1)]
reflection_type = "x"  # Change this to "y" or "xy" for different reflections

# Reflect the square
reflected_square = reflect_polygon(square, reflection_type)

# Close the shapes for plotting (return to the start point to make it a closed shape)
square.append(square[0])
reflected_square.append(reflected_square[0])

# Separate x and y coordinates
ox, oy = zip(*square)
rx, ry = zip(*reflected_square)

# Plot the original and reflected squares
plt.plot(ox, oy, color='blue', marker='o', label='Original Square')  # Original square
plt.plot(rx, ry, color='red', marker='o', label='Reflected Square')  # Reflected square
plt.title(f'2D Reflection of a Square about the {reflection_type.upper()}-axis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.axis('equal')
plt.show()
```
